```python
def eff(timed_df, type):
    mu = expected_returns.ema_historical_return(timed_df)
    mu.fillna(0, inplace=True)
    mu.replace([np.inf, -np.inf], 0, inplace=True)
    S = risk_models.exp_cov(timed_df)
    S_f = S + 1e-6 * np.eye(S.shape[0])
    ef = EfficientFrontier(mu, S_f, solver="ECOS")
    ef.add_objective(objective_functions.L2_reg, gamma=2)

    if type == "max_sharpe":
        ef.max_sharpe()

    elif type == "min":
        ef.min_volatility()

    elif type == "max":
        ef.max_quadratic_utility()

    weights = ef.clean_weights()
    portfolio_performance = ef.portfolio_performance(verbose=True)
    return weights, portfolio_performance
```

```python
def BlackLitterman(timed_df, type):
    num_days_per_year = 252

    df = timed_df

    num_stocks = len(df.columns)

    # Calculate expected returns and sample covariance matrix
    mu = expected_returns.ema_historical_return(df)
    S = risk_models.sample_cov(df)

    # Specify views
    views = {
        # "Stock_1": 0.02,  # Absolute view: 2% expected return for "Stock_1"
        # Add more views if needed
    }

    # Black-Litterman model
    bl = BlackLittermanModel(S, pi=mu, absolute_views=views)

    # Calculate posterior expected returns and covariance matrix
    bl_return = bl.bl_returns()
    bl_covariance = bl.bl_cov()

    # Optimize portfolio
    ef = EfficientFrontier(bl_return, bl_covariance, solver="ECOS")
    ef.add_objective(objective_functions.L2_reg, gamma=2)

    if type == "max_sharpe":
        ef.max_sharpe()

    elif type == "min":
        ef.min_volatility()

    elif type == "max":
        ef.max_quadratic_utility()
    cleaned_weights = ef.clean_weights()

    # Display optimal portfolio weights
    # remove zero weights
    refined_weights =   {key: value for key, value in cleaned_weights.items() if value != 0}

    portfolio_performance = ef.portfolio_performance(verbose=True)

    ef = EfficientFrontier(bl_return, bl_covariance, solver="ECOS")
    ef.add_objective(objective_functions.L2_reg, gamma=2)
    # plot the efficient frontier
  

    return refined_weights, portfolio_performance
```

```python
from tqdm.auto import tqdm
def monte_carlo(timed_df):
    num_years = 5
    num_days_per_year = 252
    num_simulations = 1_000_000

    df = timed_df
    num_stocks = len(df.columns)

    # Define portfolio weights (initial guess)
    weights = np.random.random(num_stocks)
    weights /= np.sum(weights)

    # Calculate returns and covariance matrix
    returns = df.pct_change()
    mean_returns = returns.mean()
    cov_matrix = returns.cov()

    # Monte Carlo simulation
    portfolio_returns = []
    portfolio_volatilities = []
    weights_list = []

    for _ in tqdm(range(num_simulations)):
        # Generate random weights
        weights = np.random.random(num_stocks)
        weights /= np.sum(weights)
        
        # Calculate portfolio returns and volatility
        portfolio_return = np.sum(mean_returns * weights) * num_days_per_year
        portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights))) * np.sqrt(num_days_per_year)
        
        portfolio_returns.append(portfolio_return)
        portfolio_volatilities.append(portfolio_volatility)
        weights_list.append(weights)

    portfolio_returns = np.array(portfolio_returns)
    portfolio_volatilities = np.array(portfolio_volatilities)
    weights_list = np.array(weights_list)

    return portfolio_returns, portfolio_volatilities, weights_list
```

```python
from scipy.optimize import minimize


def negative_sharpe_ratio(weights, mean_returns, cov_matrix):
    num_days_per_year = 252
    portfolio_return = np.sum(mean_returns * weights) * num_days_per_year
    portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(
        cov_matrix, weights))) * np.sqrt(num_days_per_year)
    sharpe_ratio = -portfolio_return / portfolio_volatility
    return sharpe_ratio


def optimize_portfolio(mean_returns, cov_matrix):
    num_assets = len(mean_returns)
    constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
    bounds = tuple((0, 1) for asset in range(num_assets))
    initial_guess = num_assets * [1. / num_assets,]
    optimal_weights = minimize(negative_sharpe_ratio, initial_guess, args=(mean_returns, cov_matrix,),
                               method='SLSQP', bounds=bounds, constraints=constraints)
    return optimal_weights.x


```

```python
# create the function for above
def monte_metrics(portfolio_returns, portfolio_volatilities, weights_monte):
    sharpe_ratio = portfolio_returns / portfolio_volatilities

    # find the optimal portfolio
    max_sharpe_idx = sharpe_ratio.argmax()
    sdp, rp = portfolio_volatilities[max_sharpe_idx], portfolio_returns[max_sharpe_idx]
    max_sharpe_allocation = pd.DataFrame(weights_monte[max_sharpe_idx], index=timed_df.columns, columns=['allocation'])
    max_sharpe_allocation.allocation = [round(i*100, 2) for i in max_sharpe_allocation.allocation]
    max_sharpe_allocation = max_sharpe_allocation.T

    # find the minimum volatility portfolio
    min_vol_idx = portfolio_volatilities.argmin()
    sdp_min, rp_min = portfolio_volatilities[min_vol_idx], portfolio_returns[min_vol_idx]
    min_vol_allocation = pd.DataFrame(weights_monte[min_vol_idx], index=timed_df.columns, columns=['allocation'])
    min_vol_allocation.allocation = [round(i*100, 2) for i in min_vol_allocation.allocation]
    min_vol_allocation = min_vol_allocation.T

    # find the maximum return portfolio
    max_ret_idx = portfolio_returns.argmax()
    sdp_max_ret, rp_max_ret = portfolio_volatilities[max_ret_idx], portfolio_returns[max_ret_idx]
    max_ret_allocation = pd.DataFrame(weights_monte[max_ret_idx], index=timed_df.columns, columns=['allocation'])
    max_ret_allocation.allocation = [round(i*100, 2) for i in max_ret_allocation.allocation]
    max_ret_allocation = max_ret_allocation.T

    # sdp = volatility, rp = return

    ret = {
        "max_sharpe_allocation": {
            "allocation": max_sharpe_allocation,
            "volatility": sdp,
            "return": rp,
            "sharpe_ratio": sharpe_ratio[max_sharpe_idx]
        },
        "min_vol_allocation": {
            "allocation": min_vol_allocation,
            "volatility": sdp_min,
            "return": rp_min,
            "sharpe_ratio": sharpe_ratio[min_vol_idx]
        },
        "max_ret_allocation": {
            "allocation": max_ret_allocation,
            "volatility": sdp_max_ret,
            "return": rp_max_ret,
            "sharpe_ratio": sharpe_ratio[max_ret_idx]
        }
    }

    return ret
```
